```python
import pandas as pd

class StockAnalyzer:
    def __init__(self, file_path):
        self.stock_data = pd.read_csv(file_path)

    def getMaxHighPrice(self):
        return self.stock_data['High'].max()

    def getMinLowPrice(self):
        return self.stock_data['Low'].min()

    def getAverageVolume(self):
        return self.stock_data['Volume'].mean()

    def getMaxDayGain(self):
        self.stock_data['DayGain'] = self.stock_data['Close'] - self.stock_data['Open']
        return self.stock_data['DayGain'].max()

analyzer = StockAnalyzer("./HCLTECH.NS.csv")
print(analyzer.getMaxHighPrice())
print(analyzer.getMinLowPrice())
print(analyzer.getAverageVolume())
print(analyzer.getMaxDayGain())
```
